```python
import geopandas as gpd
from shapely.geometry import LineString, mapping
from itertools import combinations

# Get a shapefile of all roads in Manhattan (New York County) from the Census Bureau
# Data source: https://www2.census.gov/geo/tiger/TIGER2020/ROADS/
df = gpd.read_file('tl_2020_36061_roads/tl_2020_36061_roads.shp')
df.head()
```

```python
df.plot(color = "black", figsize = (12, 12))
```

```python
# Check the current coordinate reference system
df['geometry'].crs
```

```python
# Transform into web mercator https://epsg.io/3857
# Projected coordinates Spherical Mercator (EPSG:3857)
# X,Y axes are in meters
df['geom_cartesian'] = df['geometry'].to_crs(3857)
df['geom_cartesian'].crs
```

```python
# Calculate lengths
df['length_meters'] = df['geom_cartesian'].length
df.head()
```

```python
# Total length of roads in Manhattan in km
df['length_meters'].sum() / 1000
```

```python
# Total length of roads in Manhattan in miles
df['length_meters'].sum() / 1609
```
